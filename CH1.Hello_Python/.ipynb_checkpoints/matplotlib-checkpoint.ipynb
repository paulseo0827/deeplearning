matplotlib

그래프를 그리기위한 라이브러리
matplotlib를 사용하면 그래프 그리기와 데이터 시각화가 쉬워짐

1) 단순한 그래프 그리기
matplotlib에 pyplot 모듈을 이용하여 sin 함수를 그리는 예제
x 값에는 arrange 메서드를 통해서[0, 0.1, 0.2, ... , 5.8, 5.9] 와 같은 데이터를 생성함
np.sin 메서드에 x 값을 할당하여 plt.plot 메서드를 호출하여 그래프를 그리고, 마지막으로 plt.show 를 호출하여 화면에 출력

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
# 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

2) pyplot 기능

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
# 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos') # 제목
plt.legend() # 도표 설명
plt.show()
```

3) 이미지 표시하기

matplotlib.image 모듈에 imread() 메서드로 이미지를 읽고, imshow() 메서드로 화면에 읽은 이미지를 출력함

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread('lena.png')

plt.imshow(img)
plt.show()
```
