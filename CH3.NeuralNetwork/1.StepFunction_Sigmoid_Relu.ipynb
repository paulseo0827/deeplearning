```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.array([-1.0, 1.0, 2.0])
print(x)
```

```python
y = x > 0
print(y)
```

```python
y = y.astype(np.int)
print(y)
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
step_y = step_function(x)
plt.plot(x, step_y)
plt.ylim(-0.1, 1.1) # y축의 범위 지정
plt.title('Step function')
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
print(sigmoid(x))
```

```python
t = np.array([1.0, 2.0, 3.0])
print(1.0 + t)
```

```python
print(1.0 / t)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
sigmoid_y = sigmoid(x)
plt.plot(x, sigmoid_y)
plt.ylim(-0.1, 1.1) # y축 범위 지정
plt.title('Sigmoid')
plt.show()
```

```python
step_y = step_function(x)
plt.plot(x, sigmoid_y, label="Sigmoid")
plt.plot(x, step_y, linestyle="--", label='StepFunction')
plt.ylim(-0.1, 1.1)
plt.title('Step function and Sigmoid')
plt.legend(loc='center right')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
relu_y = relu(x)
plt.plot(x, relu_y)
plt.ylim(-1.0, 5.5)
plt.title('Relu')
plt.show()
```
